Let's clean up the data and convert it to what we need.

Since the target variable is worth, we drop all worth=0 which is clearly wrong.

```python
dfc = df[df.worth > 0]
```

```python
dfc = dfc.drop(["name", "url", "emails", "twitter", "facebook", "instagram", "youtube", "executives", "google", "linkedin", "phones", "country"], axis=1)
```

```python
dfc.dtypes
```

```python
from sklearn import preprocessing
import copy
le = preprocessing.LabelEncoder()

mask = dfc.isnull()
dfc_prep = dfc.copy()
les = dict()
for key in dfc.keys():
  if isinstance(dfc[key][0], (str)):
    dfc_prep[key] = le.fit_transform(dfc_prep[key])
    les[key] = copy.copy(le)
```

```python
dfc_prep.where(~mask, -1, inplace=True)
```

```python
# fairly generic ml model training routine.
def train_ml(model, train_df, test_df, predict_column):
    train_X = train_df.drop(predict_column, axis=1)
    train_y = train_df[predict_column]
    test_X = test_df.drop(predict_column, axis=1)
    test_y = test_df[predict_column]
    model.fit(train_X, train_y)
    
    predict_out = model.predict(train_X)
    
    show_ml_stats('train', train_y, predict_out)
    
    predict_out = model.predict(test_X)
    
    show_ml_stats('test', test_y, predict_out)
    
    return model
```
